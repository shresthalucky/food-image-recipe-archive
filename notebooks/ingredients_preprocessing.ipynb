```python
import pandas as pd
from tqdm import tqdm

tqdm.pandas()
```

```python
det_ingrs = pd.read_csv("../dataset/more_than_2_ingredients.csv")
det_ingrs.head()
```

```python
df_train = pd.read_csv("../dataset/dataset_with_images/train_dataset_with_images.csv")
df_train.head()
```

```python
df_val = pd.read_csv("../dataset/dataset_with_images/val_dataset_with_images.csv")
df_val.head()
```

```python
df_test = pd.read_csv("../dataset/dataset_with_images/test_dataset_with_images.csv")
df_test.head()
```

```python
def filter_ingrs(df: pd.DataFrame, ref: pd.DataFrame) -> pd.DataFrame:
    """
    Replace value of ingredients of df with ingredients of ref
    """
    df = df.drop(["ingredients"], axis=1)
    df = pd.merge(df, ref, on=["id"], how="inner")

    return df
```

```python
df_train = filter_ingrs(df_train, det_ingrs)

df_train.to_csv("../dataset/dataset_valid_ingredients/train_ingrs_filteredv2.csv", index=False)
df_train.head()
```

```python
df_val = filter_ingrs(df_val, det_ingrs)

df_val.to_csv("../dataset/dataset_valid_ingredients/val_ingrs_filteredv2.csv")
df_val.head()
```

```python
df_test = filter_ingrs(df_test, det_ingrs)

df_test.to_csv("../dataset/dataset_valid_ingredients/test_ingrs_filteredv2.csv")
df_test.head()
```
