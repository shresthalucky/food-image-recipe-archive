```python
import pickle
import pandas as pd
```

```python
ingrs = pickle.load(open("../cluster_ingrs.pkl", "rb"))
```

```python
mixed_base = {}
for base, mixed in ingrs.items():
    for ing in mixed:
        mixed_base[ing] = base
```

```python
det_ingrs = pd.read_csv("../dataset/det_ingrs_2.csv")
```

```python
unrecog_ingrs = []

def extract_base_ingredients(ingrs_str):
    ingrs_list = ingrs_str.split(";")
    base_ingrs = []
    for ingr in ingrs_list:
        try:
            base_ingrs.append(mixed_base[ingr])
        except:
            unrecog_ingrs.append(ingr)
    return ";".join(set(base_ingrs))
```

```python
det_ingrs["base_ingrs"] = det_ingrs["ingredients"].map(extract_base_ingredients)
```
