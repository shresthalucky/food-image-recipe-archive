```python
def filter_col_len(source: pd.DataFrame, column: str, delim: str, length: int) -> pd.DataFrame:
  return source[source[column].map(lambda x: len(x.split(';')) > length)]
```

```python
from typing import List

stopwords = nltk.corpus.stopwords.words('english')

def lemmatize(step: str) -> List[str]:
  tokenizer = nltk.RegexpTokenizer(r'\w+')
  word_list = [w for w in tokenizer.tokenize(step) if w not in stopwords]
  return ' '.join([lemmatizer.lemmatize(w) for w in word_list])

def lemmanization_and_stopword_removal(val: str):
  steps = val.lower().split(".;")
  lemmatized_steps = map(lemmatize, steps)
  return ';'.join(lemmatized_steps)
```
