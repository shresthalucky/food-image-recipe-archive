```python
import pandas as pd
```

```python
ingrs_df = pd.read_csv("../dataset/det_ingrs_3.csv")
ingrs_df.head()
```

```python
ingrs_df = ingrs_df.drop(["ingredients"], axis=1)
```

```python
train_df = pd.read_csv("../dataset/1_dataset_with_images/train_dataset_with_images.csv")
train_df.head()
```

```python
val_df = pd.read_csv("../dataset/1_dataset_with_images/val_dataset_with_images.csv")
val_df.head()
```

```python
test_df = pd.read_csv("../dataset/1_dataset_with_images/test_dataset_with_images.csv")
test_df.head()
```

```python
def merge_dataframe(df: pd.DataFrame, ref: pd.DataFrame):
    df = df.drop(["ingredients"], axis=1)
    df = pd.merge(df, ref, on=["id"], how="inner")
    return df
```

```python
train_merged = merge_dataframe(train_df, ingrs_df)
train_merged.to_csv("../dataset/new_merged/train_merged.csv", index=False)
train_merged.head()
```

```python
val_merged = merge_dataframe(val_df, ingrs_df)
val_merged.to_csv("../dataset/new_merged/val_merged.csv", index=False)
val_merged.head()
```

```python
test_merged = merge_dataframe(test_df, ingrs_df)
test_merged.to_csv("../dataset/new_merged/test_merged.csv", index=False)
test_merged.head()
```

```python
print(f"total train => before: {len(train_df.index)} / after: {len(train_merged.index)}")
print(f"total val => before: {len(val_df.index)} / after: {len(val_merged.index)}")
print(f"total test => before: {len(test_df.index)} / after: {len(test_merged.index)}")
```
