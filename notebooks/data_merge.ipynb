```python
val_df = pd.read_csv("../dataset/dataset_lemmatized/val_lamettized.csv")
val_df.head()
```

```python
test_df = pd.read_csv("../dataset/dataset_lemmatized/test_lamettized.csv")
test_df.head()
```

```python
def merge_dataframe(df: pd.DataFrame, ref: pd.DataFrame):
    df = df.drop(["ingredients"], axis=1)
    df = pd.merge(df, ref, on=["id"], how="inner")
    return df
```

```python
# train_merged = merge_dataframe(train_df, ingrs_df)
# train_merged.to_csv("../dataset/dataset_merged/train_merged.csv", index=False)
# train_merged.head()
```

```python
val_merged = merge_dataframe(val_df, ingrs_df)
val_merged = val_merged.drop(["Unnamed: 0"], axis=1)
val_merged.to_csv("../dataset/dataset_merged/val_merged.csv", index=False)
val_merged.head()
```

```python
test_merged = merge_dataframe(test_df, ingrs_df)
test_merged = test_merged.drop(["Unnamed: 0"], axis=1)
test_merged.to_csv("../dataset/dataset_merged/test_merged.csv", index=False)
test_merged.head()
```
